```python
import cv2
import os
```

```python
video_path = "/kaggle/working/video.mp4"

cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS)

print(f"Frames per second (fps): {fps}")

cap.release()
```

```python
from tqdm import tqdm

output_path = '/kaggle/working/video_frames/'

os.makedirs(output_path, exist_ok=True)

cap = cv2.VideoCapture(video_path)

frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
progress_bar = tqdm(total=frame_count, desc='Processing frames')

frame_count = 0
while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame_count += 1

    frame_filename = output_path + f"{frame_count:04d}.jpg"
    cv2.imwrite(frame_filename, frame)
    
    progress_bar.update(1)

cap.release()
progress_bar.close()
```

```python
import zipfile

def zip_dir(directory, zip_filename):
    files = os.listdir(directory)
    
    progress = tqdm(total=len(files), desc="Zipping")
    
    with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
        
        for file in files:
            zipf.write(os.path.join(directory, file), arcname=file)
            progress.update(1)
    progress.close()

directory_to_zip = '/kaggle/working/video_frames'
zip_filename = 'video_f.zip'

zip_dir(directory_to_zip, zip_filename)
```
